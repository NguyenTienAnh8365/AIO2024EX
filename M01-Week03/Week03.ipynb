```python
import torch
import torch.nn as nn

class Softmax(nn.Module):
    def __init__(self):
        super(Softmax, self).__init__()

    def forward(self, x):
        exp_x = torch.exp(x)
        sum_exp_x = torch.sum(exp_x)
        return exp_x / sum_exp_x

class SoftmaxStable(nn.Module):
    def __init__(self):
        super(SoftmaxStable, self).__init__()

    def forward(self, x):
        c = torch.max(x)
        exp_x = torch.exp(x - c)
        sum_exp_x = torch.sum(exp_x)
        return exp_x / sum_exp_x
    
x = torch.tensor([1.0, 2.0, 3.0])
softmax = Softmax()
softmax_stable = SoftmaxStable()

print("Softmax:", softmax(x))
print("Softmax Stable:", softmax_stable(x))
```

```python
class Person:
    def __init__(self, name, yob):
        self.__name = name
        self.__yob = yob

    def describe(self):
        return f"Name: {self.__name}, Year_of_Birth: {self.__yob}"

class Student(Person):
    def __init__(self, name, yob, grade):
        super(Student, self).__init__(name, yob)
        self.__grade = grade

    def get_infor(self):
        return self._Person__name, self._Person__yob, self.__grade
    
    def get_yob(self):
        return self._Person__yob
    
    def describe(self):
        return f"{self.get_infor()}"
    
class Doctor(Person):
    def __init__(self, name, yob, specialist):
        super(Doctor, self).__init__(name, yob)
        self.__specialist = specialist

    def get_infor(self):
        return self._Person__name, self._Person__yob, self.__specialist
    
    def get_yob(self):
        return self._Person__yob
    
    def describe(self):
        return f"{self.get_infor()}"
    
class Teacher(Person):
    def __init__(self, name, yob, subject):
        super(Teacher, self).__init__(name, yob)
        self.__subject = subject

    def get_infor(self):
        return self._Person__name, self._Person__yob, self.__subject
    
    def get_yob(self):
        return self._Person__yob
    
    def describe(self):
        return f"{self.get_infor()}"

class Ward:
    def __init__(self, name):
        self.__name = name
        self.__people = []

    def add_person(self, person):
        self.__people.append(person)

    def describe(self):
        print(f"Ward Name: {self.__name}")
        for person in self.__people:
            print(person.describe())

    def count_doctor(self):
        return sum(1 for person in self.__people if isinstance(person, Doctor))

    def sort_age(self):
        sorted_people = sorted(self.__people, key=lambda person: 2024 - person.get_yob())
        return sorted_people

    def compute_average(self):
        teachers = [person for person in self.__people if isinstance(person, Teacher)]
        if not teachers:
            return 0
        avg_yob = sum(teacher.get_yob() for teacher in teachers)
        return avg_yob / len(teachers)


ward = Ward("Central Ward")

student = Student("Oanh", 2005, "1st year student")
teacher1 = Teacher("Mr. Smith", 1980, "Mathematics")
teacher2 = Teacher("Tien Anh", 2002, "English")
doctor1 = Doctor("Trang", 1990, "Nurse")
doctor2 = Doctor("Thanh", 1975, "Dentist")

ward.add_person(student)
ward.add_person(teacher1)
ward.add_person(teacher2)
ward.add_person(doctor1)
ward.add_person(doctor2)   

ward.describe()
print()

print("Number of doctors:", ward.count_doctor())

print("The arrangement increases with the year of birth:")
for sorted_person in ward.sort_age():
    print(sorted_person.describe())

print("Average year of birth of teachers:", ward.compute_average())
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack_ = []

    def is_full(self):
        return len(self.stack_) == self.capacity

    def push(self, value):
        if self.is_full():
            print(f"Stack is full, cannot push {value}!")
        else:
            self.stack_.append(value)

    def is_empty(self):
        return len(self.stack_) == 0
    
    def pop(self):
        if self.is_empty():
            print("Stack is empty, cannot pop!")
            return None
        else:
            return self.stack_.pop()

    def top(self):
        if self.is_empty():
            print("Stack is empty, no top element!")
            return None
        else:
            return self.stack_[-1]
    
    def print_stack(self):
        print("Current stack:", self.stack_)

stack = Stack(5)

stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)
stack.push(6)  

print("Stack before popping values:")
stack.print_stack()
print("Top element:", stack.top())

stack.pop()
stack.pop()

print("Stack after popping values:")
stack.print_stack()
print("Top element:", stack.top())
```

```python
class Queue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.dequeue = []

    def is_full(self):
        return len(self.dequeue) == self.capacity

    def push(self, value):
        if self.is_full():
            print(f"Queue is full, cannot push {value}!")
        else:
            self.dequeue.append(value)

    def is_empty(self):
        return len(self.dequeue) == 0
    
    def pop(self):
        if self.is_empty():
            print("Queue is empty, cannot pop!")
            return None
        else:
            return self.dequeue.pop(0)

    def top(self):
        if self.is_empty():
            print("Queue is empty, no top element!")
            return None
        else:
            return self.dequeue[0]

    def print_queue(self):
        print("Current queue:", self.dequeue)

queue = Queue(4)

queue.push(11)
queue.push(12)
queue.push(13)
queue.push(14)
queue.push(15)  

print("Queue before popping values:")
queue.print_queue()

top_element = queue.top()
print("Top element: ", top_element)

queue.pop()
queue.pop()

print("Queue after popping values:")
queue.print_queue()

top_element = queue.top()
print("Top element: ", top_element)
```
